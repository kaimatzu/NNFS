## Activation Functions
All an activation function serves is to mimic a neuron “firing” or “not firing” based on input information. Here are the types of Activation Functions, each accompanied by its respective graph (along with matplotlib code) to visually illustrate its behavior.

### The Step Activation Function
Simplest activation function. If the *weights · inputs + bias* results in a value greater than 0, the neuron will fire and output a 1; otherwise, it will output a 0.

While the step function served as a foundational element in the history of neural networks, it has several limitations that make it less popular in modern neural network architectures.
- **Not Differentiable:**
The step function is not differentiable at the point where it transitions from 0 to 1. This lack of differentiability poses challenges for gradient-based optimization algorithms like backpropagation, which rely on computing gradients to update weights during training. Modern neural networks use activation functions that are differentiable to enable efficient optimization.
- **Vanishing Gradients:**
The step function has a constant derivative of 0 almost everywhere. This leads to the vanishing gradient problem, where gradients become very small during backpropagation. In deep networks, this can result in slow or halted learning for layers closer to the input layer.
- **Binary Output:**
The step function produces binary output (0 or 1), limiting its expressive power. Modern neural networks often require activation functions that can produce a range of outputs, allowing the model to capture more complex relationships in the data.
- **Not Suitable for Regression:**
The step function is mainly used for binary classification problems. It is not suitable for regression tasks where the network needs to predict continuous values.

$$
\text{Step Function:} \\
\text{ } \\
f(x) = \begin{cases}
    0 & \text{if } x \leq 0 \\
    1 & \text{if } x > 0
\end{cases}
$$

```python
# Step Function 
import matplotlib.pyplot as plt

x = [-4, 0, 1, 4] 
y = [0, 0, 1, 1]

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Step Function')
plt.ylim(-4, 4)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.step(x, y)
plt.show()
```

### The Linear Activation Function

A linear function is simply the equation of a line. It will appear as a straight line when graphed, where y=x and the output value equals the input.

This activation function is usually applied to the last layer’s output in the case of a regression model — a model that outputs a scalar value instead of a classification. This activation function is not used in other layers due to several reasons:

- **Hidden Layers in Deep Networks:** In hidden layers of deep neural networks, using only linear activation functions may limit the expressiveness of the model. The composition of multiple linear functions is still a linear function, so the network wouldn't be able to capture complex non-linear relationships.
- **Vanishing Gradient Problem:** Same as with step activation functions, ;inear Activation functions suffer from the vanishing gradient problem. In deep networks, using only linear activations can result in diminishing gradients during backpropagation, making it difficult for the model to learn effectively.

$$
\text{Linear Function:} \\
\text{ } \\
f(x) = x
$$

```python
# Linear Function 
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4, 4)
y = np.linspace(-4, 4)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Linear Function')
plt.ylim(-4, 4)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.plot(x, y)
plt.show()
```

### The Sigmoid Activation Function

The original, more granular, activation function used for neural networks. This function returns a value that asymptotically approaches 0 for negative infinity, approaches 0.5 for an input of 0, and asymptotically approaches 1 for positive infinity.

Unlike the previous activation functions, it’s usually better to have a more granular approach for the hidden neuron activation functions. Contrary to a function like the step function, we’re getting a value that can be reversed to its original value; the returned value contains all the information from the input (e.g. an input of 3 will output the same value as an input of 300,000.)

The output from the Sigmoid function, being in the range of 0 to 1 is important for several reasons. 
- Compared to the linear function wherein the range of the output is from negative to positive infinity, the output is neatly constrained.
- The function has an S-shaped curve. This property of Sigmoid functions adds non-linearity to the output, which is crucial for the capacity of neural networks to model complex relationships and patterns in the data.

In contrast to its successor, the ReLU Activation Function, the Sigmoid Activation Function has some drawbacks:
- **Vanishing Gradient Problem:** Same as with the previous activation functions, the Sigmoid Activation function still suffers from the this problem. This is because their gradients are repeatedly multiplied during backpropagation, they can exponentially decrease or vanish as they propagate through the layers, leading to slow or stalled learning in earlier layers.
- **Not Zero-Centered:** The Sigmoid function is not zero-centered, which means that the average activation for all the neurons in a layer is not centered around zero. This can make optimization more challenging, particularly in scenarios where inputs are predominantly positive or negative.
- **Output Interpretation:** The Sigmoid function outputs values between 0 and 1, making it more suitable for binary classification problems. However, for tasks where multi-class classification or regression is required, alternative activation functions like Softmax or ReLU variants may be more effective.
- **Limited Expressiveness:** Sigmoid activation introduces less expressiveness compared to activation functions like ReLU, which allows neurons to be more active and capture a broader range of patterns. The limited range of output values may restrict the model's ability to learn complex representations.
- **Computationally Expensive:** The most important deal-breaking point.The exponential computations involved in the Sigmoid function can be computationally expensive, especially when dealing with large datasets or deep neural networks. This can affect training efficiency and overall model performance.

$$
\text{Sigmoid Function:} \\
\text{ } \\
f(x) = \frac{1}{1 + e^{-x}}
$$

```python
# Sigmoid Function 
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Sigmoid Function')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.plot(x, 1/(1+np.exp(-x)))
plt.show()
```

### Rectified Linear Activation Function

The successor of the Sigmoid Activation function. 

The Rectified Linear Units activation function (or ReLU) is y=x, clipped at 0 from the negative side. This simple yet powerful activation function is the most widely used activation function at the time of writing for various reasons — mainly speed and efficiency. The ReLU activation function maintains nonlinearity due to the bend after 0, while still being computationally cheap compared to the Sigmoid. However, this simple property is very powerful.

$$
\text{ReLU Function:} \\
\text{ } \\
f(x) = \begin{cases}
    0 & \text{if } x \leq 0 \\
    x & \text{if } x > 0
\end{cases} \\
\text{ } \\
\text{or simply} \\
\text{ } \\
f(x) = max(0,x)
$$

```python
# ReLU Function 
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('ReLU Function')
plt.ylim(-4, 4)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.plot(x, np.maximum(0, x))
plt.show()
```
